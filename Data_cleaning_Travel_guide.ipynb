### Insights

##### Table contents:
- City Column -In city column we have name of the tour package name.
- Actual_price - In this column we have the actual price of the trip without discount.
- Trip_price - In this column we have trip price after discount.
- Discount - Discount Percentage.
- Duration - No.of days and nights in the trip.

- Hotel_ratings - This column contains ratings of hotels where visitors are going to stay.

##### Modifying and adding the columns
- Creating city column:By using regular expression on visiting_places column, I extracted the first visiting city.

```python
df['city_name']=df['visiting_places'].str.extract(r'^(\w*\s*\w*\s*\w*\s*\w*\s*)')
```

```python
df.head()
```

- Renaming the city Column to Package_name to avoid confusion in city_name and city column

```python
df.rename(columns={'city': 'package_name'}, inplace=True)
```

- Package Column : In this column we don't have any duplicate or inapproprite data.
- trip_price & actual_price : In this column we don't have any duplicate or inapproprite data.Data types of these columns are object should change to integer
- Discount Column : In this column it will be better to have discount in rupees rather than percentage, so I am changing this complete column.

```python
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['trip_price'] = df['trip_price'].str.replace(',', '')
```

```python
df=df.drop(471)
```

```python
df = df.astype({'trip_price': 'int', 'actual_price': 'int'})
```

```python
df
```

```python
df=df.drop('discount', axis=1)
```

```python
df['discount']=df.actual_price-df.trip_price
```

```python
df['discount'].head()
```

```python
df.info()
```

- Duration Column - In duration column we have days and nights we can only have no.of days data, so I will remove nights and will keep days data.

```python
df['days'] = df['duration'].str.split('&').str[0].str.extract(r'(\d+)')
```

```python
df['days'].head()
```

- As we have the no.of days data we can remove the duration column

```python
df=df.drop('duration', axis=1)
```

```python
df.head()
```

- Visiting places Column - with the help of this column we can also get number of visiting places

```python
df.visiting_places=df.visiting_places.str.replace(")", "),")
```

```python
df.head()
```

```python
split_cities = df['visiting_places'].str.split(',',expand=True)
```

```python
df['no.of_visiting_cities']=split_cities.count(axis=1)-1
```

```python
df['no.of_visiting_cities']
```

- Hotel_ratings - In this column ratings are combined and clumsy without having comma in between so I am adding comma after 'r'

```python
df.hotel_ratings=df.hotel_ratings.str.replace('r', 'r,')
```

```python
df.head()
```

```python
df.info()
```

- Days column is string datatype converting to integer datatype

```python
df['days']=df['days'].astype(int)
```
