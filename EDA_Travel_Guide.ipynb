```python
def ratings():
    two_star = 0
    three_star = 0
    four_star = 0
    five_star = 0
    
    for rating in df['hotel_ratings']:
        individual_ratings = [r.strip() for r in rating.split(',')]
        
        if '2 Star' in individual_ratings:
            two_star += 1
        if '3 Star' in individual_ratings:
            three_star += 1
        if '4 Star' in individual_ratings:
            four_star += 1
        if '5 Star' in individual_ratings:
            five_star += 1

    return two_star, three_star, four_star, five_star

two_star, three_star, four_star, five_star = ratings()

print(f"2 Star: {two_star}, 3 Star: {three_star}, 4 Star: {four_star}, 5 Star: {five_star}")
```
